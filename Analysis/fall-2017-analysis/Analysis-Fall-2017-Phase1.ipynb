```python
#**Two sample unpaired t-test**

# Hypothesis: 

# About the attribute "Novelty"

# H0: mean(Full-Transparency.Novelty) = mean(Full-Transparency.Novelty)

# H1: mean(Full-Transparency.Novelty) != mean(Full-Transparency.Novelty)



attributeName = 'novelty'

# two_sample_unpaired_t_test

df_new = df1.loc[df1['hack id'] == 'unal0']
df_new2 = df1.loc[df1['hack id'] == 'unal1']
output = st.ttest_ind(df_new['novelty'], df_new2['novelty'], nan_policy='omit')
if (output.pvalue > 0.05): 
    print(output.pvalue, 'accept')
else:
    print(output.pvalue, 'reject')
```

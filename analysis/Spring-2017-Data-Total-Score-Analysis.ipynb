We can pick any value between 0 and 1 for our power, 0.8 is commonly used. 80% probability that we will correctly reject the null hypothesis. 


Effect size(d) = The estimated difference in means / pooled estimated standard deviations. 

alpha = 0.05 

power = 0.8

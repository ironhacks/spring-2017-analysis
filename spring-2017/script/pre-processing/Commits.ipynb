```python
import pandas as pd

info = pd.read_csv("../../data/processed-data/loc-win.csv")
info
```

```python
import pandas as pd

loc = pd.read_csv("../../data/processed-data/loc-final.csv")
filled = loc.fillna(0)
filled
```

```python
result = pd.merge(info, filled[['commit-id', 'files', 'additions', 'deletions']], on='commit-id', how='right')
filled_result = result.fillna(0)
final_result = filled_result[filled_result.name != 0]
final_result
```

```python
score = pd.read_csv("../scores/spring-2017-analysis/overall_score_click_result.csv")
score = score.rename(columns={'Student': 'name'})
score
```

```python
result = pd.merge(final_result, score[['name', 'Group']], on='name', how='left')
result.to_csv('result_name.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig,ax = plt.subplots()
sns.set_theme(style="whitegrid")
sns.lineplot(data=final_result, hue='name', x='date', y='additions')
sns.set(rc = {'figure.figsize':(100,5)})
#fig.set_figheight(10)
ax.set(ylim=(0, 500))
ax.set(xlim=(0, 210))
plt.legend(bbox_to_anchor=(2, 2), loc=2, borderaxespad=0.)
plt.show()
```

```python
import pandas as pd

commit = pd.read_csv("../../data/processed-data/commit-data.csv")
commit
```

```python
group1 = commit.loc[commit['Group'] == 1]
group1
```

```python
group1_additions = group1[['email_name', 'additions']]
group1_additions['group1_additions'] = group1_additions.groupby('email_name').cumsum()
group1_additions.to_csv('group1_additions')
```

```python
max_bogota1 = group1_additions.groupby(['email_name'], sort=False)['group1_additions'].max().reset_index()
max_bogota1['group1_additions'].describe()
```

```python
group3 = commit.loc[commit['Group'] == 3]
group3
```

```python
group3_additions = group3[['email_name', 'additions']]
group3_additions['group3_additions'] = group3_additions.groupby('email_name').cumsum()
group3_additions["commits_count"] = group3_additions.groupby("email_name").cumcount()+1
group3_additions
```

```python
group3_additions.to_csv("group3_additions.csv")
```

```python
group2 = commit.loc[commit['Group'] == 2]
group2
```

```python
group3_additions = group3[['email_name', 'additions']]
group3_additions['group3_additions'] = group3_additions.groupby('email_name').cumsum()
group3_additions
```

```python
max_bogota3 = group3_additions.groupby(['email_name'], sort=False)['commits_count'].max().reset_index()
max_bogota3['commits_count'].describe()
```

```python
group2_additions = group2[['email_name', 'additions']]
group2_additions['group2_additions'] = group2_additions.groupby('email_name').cumsum()
group2_additions["commits_count"] = group2_additions.groupby("email_name").cumcount()+1
group2_additions
```

```python
group2_additions.to_csv("group2_additions.csv")
```

```python
max_bogota2 = group2_additions.groupby(['email_name'], sort=False)['commits_count'].max().reset_index()
max_bogota2['commits_count'].describe()
```

```python
from scipy import stats
stats.ttest_ind(max_bogota2['commits_count'],  max_bogota3['commits_count'], equal_var = False)
```

```python
from scipy import stats
stats.ttest_ind(max_bogota1['group1_additions'],  max_bogota2['group2_additions'], equal_var = False)
```

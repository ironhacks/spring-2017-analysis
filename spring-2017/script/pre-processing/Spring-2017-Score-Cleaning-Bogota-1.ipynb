```python
import pandas as pd 

df1_phase5 = pd.read_csv("../../../../../Box/NSF IronHacks App 2016 - 2019/05 Experiments 2017/01 Hack Spring 2017/01 Summary Spring 2017/02 Data/raw-data/Absolute Score Bogota1 Phase5.csv")
df1_phase5
```

```python
df1_novelty_phase5 = df1_phase5[df1_phase5['User Requirement'] != "- - -"]
df1_novelty_phase5
```

```python
df1_final_novelty_phase5 = df1_novelty_phase5[['Student', 'User Requirement', 'Info Viz Score', 'Novelty Score']]
```

```python
df1_tech_phase5 = df1_phase5[df1_phase5['Tech Score'] != "- - -"]
df1_tech_phase5
```

```python
df1_tech_phase5[['Student', 'Tech Score']]
```

```python
df1_final_tech_phase5 = df1_tech_phase5[['Student', 'Tech Score']]
df1_final_tech_phase5
```

```python
df1_phase5 = pd.merge(df1_final_novelty_phase5, df1_final_tech_phase5, on='Student', how = 'outer')
df1_phase5
```

```python
bogota1 = pd.read_csv('../data/total1-phase5.csv')
bogota1
```

```python
bogota1 = bogota1[['Student', 'Total Score']]
bogota1
```

```python
df1_phase5 = pd.merge(df1_phase5, bogota1, on='Student', how = 'outer')
df1_phase5
```

```python
df1_phase5.insert (1, "Group", 1)
df1_phase5 = df1_phase5.fillna(0)
df1_phase5
```

```python
dict = {'User Requirement': 'user-requirement-phase5',
        'Info Viz Score': 'infovis-phase5',
        'Novelty Score': 'novelty-phase5', 
        'Tech Score': 'tech-phase5', 
        'Total Score': 'total-phase5'}
  
# call rename () method
df1_phase5.rename(columns=dict,
          inplace=True)

df1_phase5
```

```python
%store -r overall_phase1
%store -r overall_phase2
%store -r overall_phase3
%store -r overall_phase4
%store -r overall_phase5

import pandas as pd
```

```python
overall_phase5.rename(columns = {'unique':'uniquephase5'}, inplace = True)
overall_phase4.rename(columns = {'unique':'uniquephase4'}, inplace = True)
overall_phase3.rename(columns = {'unique':'uniquephase3'}, inplace = True)
overall_phase2.rename(columns = {'unique':'uniquephase2'}, inplace = True)
overall_phase1.rename(columns = {'unique':'uniquephase1'}, inplace = True)
```

```python
df45 = pd.merge(overall_phase5, overall_phase4[['Student', 'Group', 'uniquephase4']], how='left', left_on=['Student', 'Group'], right_on =['Student', 'Group'] )
df45              
```

```python
df45 = df45[df45['uniquephase5'].notna()]
df45
```

```python
df45["added45"] = df45.apply(lambda row: len(set(row.uniquephase5) - set(row.uniquephase4)), axis=1)
df45["deleted45"] = df45.apply(lambda row: len(set(row.uniquephase4) - set(row.uniquephase5)), axis=1)
df45
```

```python
df345 = pd.merge(df45, overall_phase3[['Student', 'Group', 'uniquephase3']], how='left', left_on=['Student', 'Group'], right_on =['Student', 'Group'] )
df345              
```

```python
df345["added34"] = df345.apply(lambda row: len(set(row.uniquephase4) - set(row.uniquephase3)), axis=1)
df345["deleted34"] = df345.apply(lambda row: len(set(row.uniquephase3) - set(row.uniquephase4)), axis=1)
df345
```

```python
df2345 = pd.merge(df345, overall_phase2[['Student', 'Group', 'uniquephase2']], how='left', left_on=['Student', 'Group'], right_on =['Student', 'Group'] )
df2345              
```

```python
df2345["added23"] = df2345.apply(lambda row: len(set(row.uniquephase3) - set(row.uniquephase2)), axis=1)
df2345["deleted23"] = df2345.apply(lambda row: len(set(row.uniquephase2) - set(row.uniquephase3)), axis=1)
df2345
```

```python
df12345 = pd.merge(df2345, overall_phase1[['Student', 'Group', 'uniquephase1']], how='left', left_on=['Student', 'Group'], right_on =['Student', 'Group'] )
df12345   
```

```python
df12345["added12"] = df12345.apply(lambda row: len(set(row.uniquephase2) - set(row.uniquephase1)), axis=1)
df12345["deleted12"] = df12345.apply(lambda row: len(set(row.uniquephase1) - set(row.uniquephase2)), axis=1)
df12345 = df12345.fillna(0)
df12345
```

```python
df_12 = df12345[['Student', 'Group','added12']]
df_23 = df12345[['Student', 'Group','added23']]
df_34 = df12345[['Student', 'Group','added34']]
df_45 = df12345[['Student', 'Group','added45']]

df_12.rename(columns = {'added12':'added_sum'}, inplace = True)
df_23.rename(columns = {'added23':'added_sum'}, inplace = True)
df_34.rename(columns = {'added34':'added_sum'}, inplace = True)
df_45.rename(columns = {'added45':'added_sum'}, inplace = True)

df_12['phase'] = 1
df_23['phase'] = 2
df_34['phase'] = 3
df_45['phase'] = 4
```

```python
df = df_12.append([df_23, df_34, df_45], ignore_index = True)
df
```

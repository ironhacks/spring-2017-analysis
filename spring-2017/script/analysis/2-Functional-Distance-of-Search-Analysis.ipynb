## Distance of Search 

> The set of choices available to a decision maker is discovered through search, according to Simon, 1957. Success promotes local search whereas failures promotes distance search (Billinger et. al., 2014). From Billinger et. al. (2014), we know that people start exploring from the beginning. Bodreau & Lakhani (2014) argues that transparency reduces the number of solutions tried.

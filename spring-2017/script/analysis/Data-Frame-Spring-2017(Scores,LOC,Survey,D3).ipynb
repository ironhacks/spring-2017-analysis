```python
overall_scores['improve-ur-12'] = overall_scores['user-requirement-phase2'] - overall_scores['user-requirement-phase1']
overall_scores['improve-ur-23'] = overall_scores['user-requirement-phase3'] - overall_scores['user-requirement-phase2']
overall_scores['improve-ur-34'] = overall_scores['user-requirement-phase4'] - overall_scores['user-requirement-phase3']
overall_scores['improve-ur-45'] = overall_scores['user-requirement-phase5'] - overall_scores['user-requirement-phase4'] 
overall_scores['improve-ur-15'] = overall_scores['user-requirement-phase5'] - overall_scores['user-requirement-phase1']
overall_scores['improve-ur-25'] = overall_scores['user-requirement-phase5'] - overall_scores['user-requirement-phase2']
overall_scores['improve-vis-12'] = overall_scores['infovis-phase2'] - overall_scores['infovis-phase1']
overall_scores['improve-vis-23'] = overall_scores['infovis-phase3'] - overall_scores['infovis-phase2']
overall_scores['improve-vis-34'] = overall_scores['infovis-phase4'] - overall_scores['infovis-phase3']
overall_scores['improve-vis-45'] = overall_scores['infovis-phase5'] - overall_scores['infovis-phase4']
overall_scores['improve-vis-15'] = overall_scores['infovis-phase5'] - overall_scores['infovis-phase1']
overall_scores['improve-vis-25'] = overall_scores['infovis-phase5'] - overall_scores['infovis-phase2']
overall_scores['improve-novel-12'] = overall_scores['novelty-phase2'] - overall_scores['novelty-phase1']
overall_scores['improve-novel-23'] = overall_scores['novelty-phase3'] - overall_scores['novelty-phase2']
overall_scores['improve-novel-34'] = overall_scores['novelty-phase4'] - overall_scores['novelty-phase3']
overall_scores['improve-novel-45'] = overall_scores['novelty-phase5'] - overall_scores['novelty-phase4']
overall_scores['improve-novel-15'] = overall_scores['novelty-phase5'] - overall_scores['novelty-phase1']
overall_scores['improve-novel-25'] = overall_scores['novelty-phase5'] - overall_scores['novelty-phase2']
overall_scores['improve-tech-12'] = overall_scores['tech-phase2'] - overall_scores['tech-phase1']
overall_scores['improve-tech-23'] = overall_scores['tech-phase3'] - overall_scores['tech-phase2']
overall_scores['improve-tech-34'] = overall_scores['tech-phase4'] - overall_scores['tech-phase3']
overall_scores['improve-tech-45'] = overall_scores['tech-phase5'] - overall_scores['tech-phase4']
overall_scores['improve-tech-25'] = overall_scores['tech-phase5'] - overall_scores['tech-phase2']
overall_scores['improve-tech-15'] = overall_scores['tech-phase5'] - overall_scores['tech-phase1']
overall_scores['improve-total-12'] = overall_scores['total-phase2'] - overall_scores['total-phase1']
overall_scores['improve-total-23'] = overall_scores['total-phase3'] - overall_scores['total-phase2']
overall_scores['improve-total-34'] = overall_scores['total-phase4'] - overall_scores['total-phase3']
overall_scores['improve-total-45'] = overall_scores['total-phase5'] - overall_scores['total-phase4']
overall_scores['improve-total-15'] = overall_scores['total-phase5'] - overall_scores['total-phase1']
overall_scores['improve-total-25'] = overall_scores['total-phase5'] - overall_scores['total-phase2']

overall_scores
```

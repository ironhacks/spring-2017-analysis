```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from statsmodels.api import add_constant
from statsmodels.sandbox.regression.gmm import IV2SLS

%store -r novelty


df = pd.read_csv("../../../../Library/CloudStorage/Box-Box/NSF IronHacks App 2016 - 2019/05 Experiments 2017/01 Hack Spring 2017/01 Summary Spring 2017/02 Data/processed-data/overall_score_click_result.csv")
df
```

```python
df = df[(df['Group'] == 2) | (df['Group'] == 3)]
df
```

```python
df = pd.merge(df, novelty[['Student', 'Q7_Q7_1', 'Q7_Q7_2', 'Q8_Q8_1', 'Q10']], how='left', left_on=['Student'], right_on =['Student']) 
df
```

```python
#Compare the performance of 2SLS with OLS of ln(wage) on performance distance to the best
df['ln_add_loc15'] =  np.log(df['add-loc-15'] + 1)

reg_expr = 'ln_add_loc15 ~ Group + Q7_Q7_1 + Q10'
olsr_model = smf.ols(formula=reg_expr, data=df)
olsr_model_results = olsr_model.fit()
print(olsr_model_results.summary())
```

```python
g3 = df[df['Group'] == 3]
g3['ln_add_loc15'].mean()
```

```python
g2 = df[df['Group'] == 2]
g2['ln_add_loc15'].mean()
```

```python
#Compare the performance of 2SLS with OLS of ln(wage) on performance distance to the best
df['ln_total_phase5'] =  np.log(df['total-phase5'] + 1)

reg_expr = 'ln_total_phase5 ~ Group'
olsr_model = smf.ols(formula=reg_expr, data=df)
olsr_model_results = olsr_model.fit()
print(olsr_model_results.summary())
```

```python
#Compare the performance of 2SLS with OLS of ln(wage) on performance distance to the best
df['ln_infovis'] =  np.log(df['infovis-phase5'] + 1)

reg_expr = 'ln_infovis ~ Group'
olsr_model = smf.ols(formula=reg_expr, data=df)
olsr_model_results = olsr_model.fit()
print(olsr_model_results.summary())
```

```python
df['project_clicks_hacker'] = df['project-clicks-hacker'].fillna(0)
df['ln_project'] =  np.log(df['project-clicks-hacker'] + 1)

reg_expr = 'ln_project ~ Group'
olsr_model = smf.ols(formula=reg_expr, data=df)
olsr_model_results = olsr_model.fit()
print(olsr_model_results.summary())
```

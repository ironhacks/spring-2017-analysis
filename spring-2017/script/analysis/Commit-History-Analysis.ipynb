```python
%store -r overall_loc_scores

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

overall_loc_scores
```

```python
overall_loc_scores.dropna(subset = ["Group"], inplace=True)
overall_loc_scores
```

```python
commit = pd.read_csv('../../../../Library/CloudStorage/Box-Box/NSF IronHacks App 2016 - 2019/05 Experiments 2017/01 Hack Spring 2017/01 Summary Spring 2017/02 Data/processed-data/final_version_commit.csv')
commit
```

```python
df_remove = commit.dropna(subset=['phase'])
df_remove
```

```python
df_remove = df_remove.rename(columns={"username": "Student"})
df = pd.merge(overall_loc_scores, df_remove,  how='left', left_on=['Student','Group'], right_on = ['Student','Group'])
df
```

```python
df['Group'].isnull().sum()
```

```python
group3 = df[df['Group'] == 3]
group2 = df[df['Group'] == 2]
group1 = df[df['Group'] == 1]
group0 = df[df['Group'] == 0]
```

## Trial Lines Added 

```python
df_add_median_group0 = []
df_add_mean_group0 = []
for i in range(1, 6):
    phase = group0[group0['phase'] == i]
    df_new = phase.groupby('Student')['additions'].sum()
    plt.figure()
    plt.title("Distributions of Trial Lines Added for Bogota 0")
    sns.distplot(np.log(df_new+1))
    df_add_median_group0.append(np.log(df_new+1).median())
    df_add_mean_group0.append(np.log(df_new+1).mean())

df_add_median_group1 = []
df_add_mean_group1 = []
for i in range(1, 6):
    phase = group1[group1['phase'] == i]
    df_new = phase.groupby('Student')['additions'].sum()
    plt.figure()
    plt.title("Distributions of Trial Lines Added for Bogota 1")
    sns.distplot(np.log(df_new+1))
    df_add_median_group1.append(np.log(df_new+1).median())
    df_add_mean_group1.append(np.log(df_new+1).mean())

    
df_add_median_group2 = []
df_add_mean_group2 = []
for i in range(1, 6):
    phase = group2[group2['phase'] == i]
    df_new = phase.groupby('Student')['additions'].sum()
    plt.figure()
    plt.title("Distributions of Trial Lines Added for Bogota 2")
    sns.distplot(np.log(df_new+1))
    df_add_median_group2.append(np.log(df_new+1).median())
    df_add_mean_group2.append(np.log(df_new+1).mean())


df_add_median_group3 = []
df_add_mean_group3 = []
for i in range(1, 6):
    phase = group3[group3['phase'] == i]
    df_new = phase.groupby('Student')['additions'].sum()
    plt.figure()
    plt.title("Distributions of Trial Lines Added for Bogota 3")
    sns.distplot(np.log(df_new+1))
    df_add_median_group3.append(np.log(df_new+1).median())
    df_add_mean_group3.append(np.log(df_new+1).mean())
```

```python
import matplotlib.pyplot as plt

# Mean Scores of All Groups [ Tech Scores ]

y1 = df_add_median_group0
y2 = df_add_median_group1
y3 = df_add_median_group2
y4 = df_add_median_group3

x_axis=['Phase 1', 'Phase 2', 'Phase 3', 'Phase 4', 'Phase 5']

line1 = plt.plot(x_axis, y1,'bo-',label='Bogota 0 ( No Transparency )') 
line2 = plt.plot(x_axis, y2,'go-',label='Bogota 1 ( Performance Transparency )') 
line3 = plt.plot(x_axis, y3,'ko-',label='Bogota 2 ( Solution Transparency )')
line4 = plt.plot(x_axis, y4,'ro-',label='Bogota 3 ( Full Transparency )') 

plt.title("Median Trial Lines Added Per Individual")
plt.ylabel("Median trial Lines Added Per Individual")

plt.legend(bbox_to_anchor=(1.5, 1),
           bbox_transform=plt.gcf().transFigure)
```

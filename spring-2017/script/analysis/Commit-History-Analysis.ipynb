## Commit History


In this notebook, we are examining the commit history across the timelines. This is not a cumulative measure of all lines of codes added since the start of the contest. For example,  We have 2 measures here, lines of codes added and deleted. 

* Experiential search is characterized as a process of local search (Gavetti & Levinthal, 2011). 

* Experiential learning tends to consist of local, hill-climbing processes. the search over the space of alternatives in the neighborhood of current practices ( Gavetti & Levinthal, 2011 )

* Cognitive search is forward looking whereas experiential search is backward-looking (Gavetti & Levinthal, 2011). 

* They differ in their extentiveness of choices. (Gavetti & Levinthal, 2011). 

* Experiential process implies that only one alternative at a time may be explored,such that alternatives are explored sequentially.In contrast, an off-line, cognitive process may invoke a broadset of alternative actions.  (Gavetti & Levinthal, 2011). 

* Currently not clear whether it's an experiential search.

## T-test for Trial Lines Added

```python
# T-tests between Bogota 2 and Bogota 1

import pandas as pd
import inspect
from scipy import stats

def retrieve_name(var):
   callers_local_vars = inspect.currentframe().f_back.f_globals.items()
   return [var_name for var_name, var_val in callers_local_vars if var_val is var][0]

list1,list2 = [group2_phase1['additions'], group2_phase2['additions'], group2_phase3['additions'], group2_phase4['additions'], group2_phase5['additions']],[group3_phase1['additions'], group3_phase2['additions'], group3_phase3['additions'], group3_phase4['additions'], group3_phase5['additions']] 
for x in range(len(list1)):
    for y in range(len(list2)):
        print('---------')
        print(stats.ttest_ind(list1[x], list2[y]))
        
```

## Rare Functions Match for Phase 1

```python
remove_0 = overall_phase1.loc[(overall_phase1['len_unique'] != 0 )]
remove_na = remove_0.dropna()
remove_na
```

```python
type(overall_phase1['unique'][3])
```

```python
selected_10 = df_all.loc[(df_all['Count'] == 1 ) | (df_all['Count'] == 2 )]
core_list = selected_10['Functions'].to_list()
core_list
```

```python
def matchFunctions(row):
    result = []
    for func in row['unique']:
        if func in core_list:
            result.append(func)
    return result if len(result) > 0 else 0

remove_na['match'] = remove_na.apply(matchFunctions, axis = 1)
remove_na
```

```python
remove_na['len_match'] = remove_na['match'].str.len()
remove_na_phase1 = remove_na[['len_match', 'Group']]
print(remove_na_phase1.groupby('Group').describe().unstack(1))
```

```python
remove_na_phase1['len_match'].hist(by=remove_na_phase1['Group'])
```

```python
pd.set_option('display.max_columns', None)
remove_na_phase1.groupby('Group')['len_match'].value_counts().unstack(fill_value=0)
```

```python
median_log_bogota3 = []
median_log_bogota2 = []
median_log_bogota1 = []
median_log_bogota0 = []

mean_log_bogota3 = []
mean_log_bogota2 = []
mean_log_bogota1 = []
mean_log_bogota0 = []

max_log_bogota3 = []
max_log_bogota2 = []
max_log_bogota1 = []
max_log_bogota0 = []

for df in [overall_phase1_bogota3, overall_phase2_bogota3, overall_phase3_bogota3, overall_phase4_bogota3, overall_phase5_bogota3]:
    median_value = (np.log(df['len_unique_library']+1)).median()
    mean_value = (np.log(df['len_unique_library']+1)).mean()
    max_value = (np.log(df['len_unique_library']+1)).max()
    max_log_bogota3.append(max_value)
    median_log_bogota3.append(median_value)
    mean_log_bogota3.append(mean_value)
    plt.figure()
    plt.title("Log Transform Library Count for Bogota 3")
    sns.distplot(np.log(df['len_unique_library']+1))

for df in [overall_phase1_bogota2, overall_phase2_bogota2, overall_phase3_bogota2, overall_phase4_bogota2, overall_phase5_bogota2]:
    median_value = (np.log(df['len_unique_library']+1)).median()
    mean_value = (np.log(df['len_unique_library']+1)).mean()
    max_value = (np.log(df['len_unique_library']+1)).max()
    max_log_bogota2.append(max_value)
    median_log_bogota2.append(median_value)
    mean_log_bogota2.append(mean_value)
    plt.figure()
    plt.title("Log Transform Library Count for Bogota 2")
    sns.distplot(np.log(df['len_unique_library']+1))
    
for df in [overall_phase1_bogota1, overall_phase2_bogota1, overall_phase3_bogota1, overall_phase4_bogota1, overall_phase5_bogota1]:
    median_value = (np.log(df['len_unique_library']+1)).median()
    mean_value = (np.log(df['len_unique_library']+1)).mean()
    max_value = (np.log(df['len_unique_library']+1)).max()
    max_log_bogota1.append(max_value)
    median_log_bogota1.append(median_value)
    mean_log_bogota1.append(mean_value)
    plt.figure()
    plt.title("Log Transform Library Count for Bogota 1")
    sns.distplot(np.log(df['len_unique_library']+1))

for df in [overall_phase1_bogota0, overall_phase2_bogota0, overall_phase3_bogota0, overall_phase4_bogota0, overall_phase5_bogota0]:
    median_value = (np.log(df['len_unique_library']+1)).median()
    mean_value = (np.log(df['len_unique_library']+1)).mean()
    max_value = (np.log(df['len_unique_library']+1)).max()
    max_log_bogota0.append(max_value)
    median_log_bogota0.append(median_value)
    mean_log_bogota0.append(mean_value)
    plt.figure()
    plt.title("Log Transform Library Count for Bogota 0")
    sns.distplot(np.log(df['len_unique_library']+1))
```

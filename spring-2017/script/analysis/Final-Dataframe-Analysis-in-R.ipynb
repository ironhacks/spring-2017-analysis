```python
%store -r exploration
%store -r variety

exploration['phase'] = exploration['phase'] - 1
exploration = exploration.rename(columns={'user2': 'Student'})

variety['phase'] = variety['phase'] - 1
```

```python
import pandas as pd
df = pd.read_csv("scores_commits.csv")
df = df.rename(columns={"novelty": "novelty_score"})
df = df[df['phase'] != 1]
df['phase'] = df['phase'] - 1
df
```

```python
df1 = pd.read_csv("novelty_complete_15_12.csv")
df1
```

```python
df_merged = pd.merge(df1, df[['Student', 'phase', 'count', 'total', 'user-requirement', 'infovis', 'novelty_score']], how='left')
df_merged
```

```python
df_merged = pd.merge(df_merged, exploration[['Student', 'group', 'phase', 'exploration']], how='left')
df_merged
```

```python
pd.set_option('display.max_rows', None)
df_merged['exploration'] = df_merged['exploration'].fillna(0)
df_merged
```

```python
df_merged = pd.merge(df_merged, variety[['Student', 'phase', 'Group', 'len_unique']], how='left')
df_merged
```

```python
df_merged.to_csv('merged.csv')
```

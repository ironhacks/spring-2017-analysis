```python
from scipy import stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import networkx as nx
np.random.seed(0)
```

```python
df = pd.read_csv("all_score_phase5_update.csv")
df1 = df.dropna()
df1.head(5)
```

```python
honors0 = df1.loc[df1['hack id'] == 'honors0']
#print(honors0)
result0 = honors0[['hack id', 'novelty']]
result0_revised = pd.DataFrame(result0)

unal0 = df1.loc[df1['hack id'] == 'unal0']
result1 = unal0[['hack id', 'novelty']]
result1_revised = pd.DataFrame(result1)

unal1 = df1.loc[df1['hack id'] == 'unal1']
result2 = unal1[['hack id', 'novelty']]
result2_revised = pd.DataFrame(result2)

unal2 = df1.loc[df1['hack id'] == 'unal2']
result3 = unal2[['hack id', 'novelty']]
result3_revised = pd.DataFrame(result3)

append1 = result0_revised.append(result1_revised)
append2 = append1.append(result2_revised)
append3 = append2.append(result3_revised)

append3.columns = ['treatments', 'value']
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Ordinary Least Squares (OLS) model
model = ols('value ~ C(treatments)', data=append3).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
honors0 = df1.loc[df1['hack id'] == 'honors0']
#print(honors0)
result0 = honors0['novelty']
result0_revised = pd.DataFrame(result0)

unal0 = df1.loc[df1['hack id'] == 'unal0']
result1 = unal0['novelty']
result1_revised = pd.DataFrame(result1)

unal1 = df1.loc[df1['hack id'] == 'unal1']
result2 = unal1['novelty']
result2_revised = pd.DataFrame(result2)

unal2 = df1.loc[df1['hack id'] == 'unal2']
result3 = unal2['novelty']
result3_revised = pd.DataFrame(result3)

fvalue, pvalue = stats.f_oneway(result0_revised['novelty'], result1_revised['novelty'], result2_revised['novelty'], result3_revised['novelty'])
print(fvalue, pvalue)
```

```python
honors0 = df1.loc[df1['hack id'] == 'honors0']
#print(honors0)
result0 = honors0['novelty_normal']
result0_revised = pd.DataFrame(result0)

unal0 = df1.loc[df1['hack id'] == 'unal0']
result1 = unal0['novelty_normal']
result1_revised = pd.DataFrame(result1)

unal1 = df1.loc[df1['hack id'] == 'unal1']
result2 = unal1['novelty_normal']
result2_revised = pd.DataFrame(result2)

unal2 = df1.loc[df1['hack id'] == 'unal2']
result3 = unal2['novelty_normal']
result3_revised = pd.DataFrame(result3)

fvalue, pvalue = stats.f_oneway(result0_revised['novelty_normal'], result1_revised['novelty_normal'], result2_revised['novelty_normal'], result3_revised['novelty_normal'])
print(fvalue, pvalue)
```

# Interpretation
## The p value obtained from ANOVA analysis is significant (p < 0.05), and therefore, we conclude that there are significant differences among treatments.

## Note on F value: F value is inversely related to p value and higher F value (greater than F critical value) indicates a significant p value.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# perform multiple pairwise comparison (Tukey HSD)
m_comp = pairwise_tukeyhsd(endog=append3['value'], groups=append3['treatments'], alpha=0.05)
print(m_comp)
```

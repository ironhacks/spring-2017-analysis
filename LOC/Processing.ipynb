# Phase 4 - 5 ( Code Added, Deleted, Number of Files Changed )

```python
import pandas as pd

df = pd.read_csv('phase-4-5-prelim.csv')
df= df.drop(['Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6'], axis = 1)
```

```python
df['change-revised'] = df.change.str.extract('(\d+)')
df.head(5)
```

```python
df['add-revised'] = df.addition.str.extract('(\d+)')
df['add-revised'] = df['add-revised'].fillna(0)
df.head(5)
```

```python
df['delete-revised'] = df.delete.str.extract('(\d+)')
df['delete-revised'] = df['delete-revised'].fillna(0)
df.head(5)
```

# Phase 4 -5 - Filter Bogota 3 ( Full Transparency Group )

```python
df_name = pd.read_csv('bogota-3.csv')
df_name.head(5)
```

```python
df['bogota-3'] = df['username'].isin(df_name['Project'])
df.head(5)
```

```python
df["bogota-3"] = df["bogota-3"].astype(int)
df.head(5)
```

```python
df_new = df.loc[df['bogota-3'] == 1]
df_new.head(5)
```

```python
df_new['change-revised'] = pd.to_numeric(df_new['change-revised'])
print(df_new['change-revised'].mean())
print(df_new['change-revised'].median())
print(df_new['change-revised'].std())
print(df_new['change-revised'].max())
print(df_new['change-revised'].min())
```

```python
df_new['delete-revised'] = pd.to_numeric(df_new['delete-revised'])
df_new['delete-revised'].mean()
print(df_new['delete-revised'].mean())
print(df_new['delete-revised'].median())
print(df_new['delete-revised'].std())
print(df_new['delete-revised'].max())
print(df_new['delete-revised'].min())

df3_delete = df_new['delete-revised']
df3_delete.values
```

```python
df_new['add-revised'] = pd.to_numeric(df_new['add-revised'])
df_new['add-revised'].mean()

print(df_new['add-revised'].mean())
print(df_new['add-revised'].median())
print(df_new['add-revised'].std())
print(df_new['add-revised'].max())
print(df_new['add-revised'].min())
df3_add = df_new['add-revised']
df3_add.values
```

```python
df_new['add-revised'] = df_new['add-revised'].astype(float)
df_new['delete-revised'] = df_new['delete-revised'].astype(float)
df_new['effort'] = df_new['add-revised'] - 0.5 * df_new['delete-revised']
print(df_new['effort'].mean())
print(df_new['effort'].median())
print(df_new['effort'].std())
print(df_new['effort'].max())
print(df_new['effort'].min())
```

```python
import numpy as np
np.set_printoptions(suppress=True)
diff3 = df_new['add-revised'] - 0.5 * df_new['delete-revised']
diff3.values
```
